```python
import spacy
import pandas as pd
import numpy as np
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
text = 'Rahul is a very interesting character to talk to. We can have a long conversation about physics and astronomy'
doc = nlp(text)
stopwords = []
for word in doc:
  if word.is_stop:
    stopwords.append(word)

stopwords
```

```python
def removestopword(text):
  doc = nlp(text)
  rmstopwords = []
  for word in doc:
    if not word.is_stop:
      rmstopwords.append(word.text)

  return " ".join(rmstopwords)
```

```python
text = 'Rahul is a very interesting character to talk to. We can have a long conversation about physics and astronomy'
catch = removestopword(text)
catch
```
